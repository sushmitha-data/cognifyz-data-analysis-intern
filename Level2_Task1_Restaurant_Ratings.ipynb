```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")
print("=" * 50)
print("RATINGS ANALYSIS")
print("=" * 50)
rating_counts = df['Aggregate rating'].value_counts().sort_index()
most_common_rating = df['Aggregate rating'].mode()[0]
print(f"Most common rating: {most_common_rating}")
avg_votes = df['Votes'].mean()
print(f"Average votes per restaurant: {avg_votes:.2f}")
print(f"Highest votes: {df['Votes'].max()}")
print(f"Lowest votes: {df['Votes'].min()}")
plt.figure(figsize=(10, 6))
plt.hist(df['Aggregate rating'], bins=20,color='steelblue', edgecolor='black')
plt.title("Distribution of Restaurant Ratings",  fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Aggregate Rating", fontsize=12, color='darkgreen')
plt.ylabel("Number of Restaurants", fontsize=12, color='darkgreen')
plt.tight_layout()
plt.savefig("Level2_Task1_Ratings.png")
plt.show()
plt.figure(figsize=(10, 6))
rating_groups = df.groupby('Aggregate rating')['Votes'].mean()
plt.bar(rating_groups.index, rating_groups.values,color='coral', edgecolor='black')
plt.title("Average Votes by Rating",fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Aggregate Rating", fontsize=12, color='darkgreen')
plt.ylabel("Average Votes", fontsize=12, color='darkgreen')
plt.tight_layout()
plt.savefig("Level2_Task1_Votes.png")
plt.show()
print("Charts saved!")
```
