```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
df = pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")
print("GEOGRAPHIC ANALYSIS")
df_map = df.dropna(subset=['Latitude', 'Longitude'])
df_map = df_map[(df_map['Latitude'] != 0) & (df_map['Longitude'] != 0)]
print(f"Total restaurants with location: {len(df_map)}")
print(f"Latitude range: {df_map['Latitude'].min():.2f} to {df_map['Latitude'].max():.2f}")
print(f"Longitude range: {df_map['Longitude'].min():.2f} to {df_map['Longitude'].max():.2f}")
restaurant_map = folium.Map(
    location=[df_map['Latitude'].mean(), df_map['Longitude'].mean()],
    zoom_start=4)
heat_data = [[row['Latitude'], row['Longitude']]
             for index, row in df_map.iterrows()]
HeatMap(heat_data).add_to(restaurant_map)
restaurant_map.save("Level2_Task3_Geographic_Map.html")
print("Map saved as HTML file!")
plt.figure(figsize=(12, 8))
plt.scatter(df_map['Longitude'], df_map['Latitude'],
            c=df_map['Aggregate rating'],
            cmap='YlOrRd', alpha=0.5, s=10)
plt.colorbar(label='Rating')
plt.title("Restaurant Locations Worldwide",
          fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Longitude", fontsize=12, color='darkgreen')
plt.ylabel("Latitude", fontsize=12, color='darkgreen')
plt.tight_layout()
plt.savefig("Level2_Task3_Geographic.png")
plt.show()
print("Charts saved!")
```
