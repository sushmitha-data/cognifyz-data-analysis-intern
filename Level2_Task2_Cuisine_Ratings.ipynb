```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")
print("CUISINE COMBINATION ANALYSIS")
cuisine_counts = df['Cuisines'].value_counts().head(10)
print("\nTop 10 Common Cuisine Combinations:")
for cuisine, count in cuisine_counts.items():
    print(f"{cuisine}: {count} restaurants")
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean()
top_cuisine_ratings = cuisine_ratings[cuisine_counts.index]
print("\n" + "=" * 50)
print("AVERAGE RATINGS FOR TOP COMBINATIONS")
print("=" * 50)
for cuisine, rating in top_cuisine_ratings.items():
    print(f"{cuisine}: {rating:.2f} average rating")
plt.figure(figsize=(12, 6))
plt.bar(range(len(cuisine_counts)), cuisine_counts.values,
        color='red', edgecolor='black')
plt.title("Top 10 Most Common Cuisine Combinations",
          fontsize=16, fontweight='bold', color='red')
plt.xlabel("Cuisine Combination", fontsize=12, color='darkgreen')
plt.ylabel("Number of Restaurants", fontsize=12, color='darkgreen')
plt.xticks(range(len(cuisine_counts)),
           [c[:20] + '...' if len(c) > 20 else c
            for c in cuisine_counts.index],
           rotation=45, ha='right')
plt.tight_layout()
plt.savefig("Level2_Task2_Cuisine_Combination.png")
plt.show()
plt.figure(figsize=(12, 6))
plt.bar(range(len(top_cuisine_ratings)),
        top_cuisine_ratings.values,
        color='yellow', edgecolor='black')
plt.title("Average Ratings for Top Cuisine Combinations",
          fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Cuisine Combination", fontsize=12, color='darkgreen')
plt.ylabel("Average Rating", fontsize=12, color='darkgreen')
plt.xticks(range(len(top_cuisine_ratings)),
           [c[:20] + '...' if len(c) > 20 else c
            for c in top_cuisine_ratings.index],
           rotation=45, ha='right')
plt.ylim(0, 5)
plt.tight_layout()
plt.savefig("Level2_Task2_Cuisine_Ratings.png")
plt.show()
print("Charts saved!")
```
