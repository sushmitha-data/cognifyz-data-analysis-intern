```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")
print("=" * 50)
print("RESTAURANT CHAINS ANALYSIS")
print("=" * 50)
restaurant_counts = df['Restaurant Name'].value_counts()
chains = restaurant_counts[restaurant_counts > 1]
print(f"\nTotal unique restaurants: {len(restaurant_counts)}")
print(f"Total restaurant chains: {len(chains)}")
print(f"\nTop 10 Restaurant Chains:")
print(chains.head(10))
top_chains = chains.head(10).index
chain_ratings = df[df['Restaurant Name'].isin(top_chains)].groupby('Restaurant Name')['Aggregate rating'].mean()
print("\n" + "=" * 50)
print("AVERAGE RATINGS FOR TOP CHAINS")
print("=" * 50)
for chain, rating in chain_ratings.sort_values(ascending=False).items():
    print(f"{chain}: {rating:.2f} average rating")
plt.figure(figsize=(12, 6))
plt.bar(range(len(chains.head(10))),
        chains.head(10).values,
        color='steelblue', edgecolor='black')
plt.title("Top 10 Restaurant Chains by Number of Locations",
          fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Restaurant Chain", fontsize=12, color='darkgreen')
plt.ylabel("Number of Locations", fontsize=12, color='darkgreen')
plt.xticks(range(len(chains.head(10))),
           chains.head(10).index,
           rotation=45, ha='right')
plt.tight_layout()
plt.savefig("Level2_Task4_Chains.png")
plt.show()
plt.figure(figsize=(12, 6))
chain_ratings_sorted = chain_ratings.sort_values(ascending=False)
plt.bar(range(len(chain_ratings_sorted)),
        chain_ratings_sorted.values,
        color='coral', edgecolor='black')
plt.title("Average Ratings for Top Restaurant Chains",
          fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Restaurant Chain", fontsize=12, color='darkgreen')
plt.ylabel("Average Rating", fontsize=12, color='darkgreen')
plt.xticks(range(len(chain_ratings_sorted)),
           chain_ratings_sorted.index,
           rotation=45, ha='right')
plt.ylim(0, 5)
plt.tight_layout()
plt.savefig("Level2_Task4_Chain_Ratings.png")
plt.show()
print("Charts saved!")
```
