```python
import pandas as pd
import matplotlib.pyplot as plt 
df=pd.read_csv(r"C:\Users\SUSHMITHA RAJU\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\Dataset._csv.csv")
delivery_counts=df["Has Online delivery"].value_counts()
total =len(df)
print('='*50)
print("online deliver")
print("="*50)
for delivery,count in delivery_counts.items():
    percentage=(count/total)*100
print(f"{delivery}:{count}restaurants({percentage:.2f}%)")
avg_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print("\n" + "=" * 50)
print("AVERAGE RATINGS")
print("=" * 50)
for delivery, rating in avg_ratings.items():
    print(f"{delivery}: {rating:.2f} average rating")

plt.figure(figsize=(8, 5))
plt.bar(delivery_counts.index, delivery_counts.values,
        color=['purple', 'gold'],
        edgecolor='black',
        linewidth=1.5)
plt.title("Online Delivery Distribution",
          fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Has Online Delivery", fontsize=12, color='darkgreen')
plt.ylabel("Number of Restaurants", fontsize=12, color='darkgreen')
plt.tight_layout()
plt.savefig("Level1_Task4_Online_Delivery.png")
plt.show()

# Bar chart 2 - Average ratings comparison
plt.figure(figsize=(8, 5))
plt.bar(avg_ratings.index, avg_ratings.values,
        color=['coral', 'turquoise'],
        edgecolor='black',
        linewidth=1.5)
plt.title("Average Ratings: With vs Without Online Delivery",
          fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Has Online Delivery", fontsize=12, color='darkgreen')
plt.ylabel("Average Rating", fontsize=12, color='darkgreen')
plt.ylim(0, 5)
plt.tight_layout()
plt.savefig("Level1_Task4_Ratings_Comparison.png")
plt.show()
print("Charts saved!")



```
